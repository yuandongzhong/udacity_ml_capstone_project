```python
from keras.preprocessing import image
from helper import log_progress
import numpy as np
import os
import pickle


# we create a helper function to detect and create a list of the 'incorrect' images, 
# by making use of the pre-trained network/model

def get_wrong_pic_list(model, preprocess_input, decode_predictions, top_number=50):    
    pic_list = [] 

    # loop through all the files and make predictions
    for root, dirs, files in os.walk("train/", topdown=True):
        
        for name in log_progress(files, name="Progress"):       # using a progress bar to track the progress
            if name[0] == 'c' or name[0] == 'd':                # exclude system hidden file like '.DS_Store' 
                image_path = root + "/" + name
                img = image.load_img(image_path, target_size=(224, 224)) 
                x = image.img_to_array(img)
                x = np.expand_dims(x, axis=0)
                x = preprocess_input(x)
                preds = model.predict(x)
                decode_preds = decode_predictions(preds, top=top_number)[0]
                is_good_picture = False                         # assume it is a wrong picture

                # loop through all predictions and
                # check if the predicition contains cats or dogs
                for i in decode_preds:
                    if not is_good_picture: 
                        if i[0] in dogs or i[0] in cats:
                            is_good_picture = True
                    else:
                        # break the loop if cat or dog is found
                        break

                # if there is no dog or cat in the picture, then it is a bad picture
                if not is_good_picture:
                    pic_list.append(image_path)
                    
    print(str(len(pic_list)) + " wrong pictures are found!")                
    return pic_list
```

### Using Xception to detect 'incorrect' images

```python
from keras.applications.xception import Xception
from keras.applications.xception import preprocess_input, decode_predictions

model = Xception(weights='imagenet')
bad_picture_list = get_wrong_pic_list(model,preprocess_input, decode_predictions)

# save the list to a file 
with open('bad_pic_list_1', 'wb') as fp:
    pickle.dump(bad_picture_list, fp)

```

# Data splitting

Here we are going to programmatically split the data into train set and validation

```python
import os
import shutil

# create dir of all the folders
train_cat_folder = "data/train/cats/"
validation_cat_folder = "data/validation/cats/"
train_dog_folder = "data/train/dogs/"
validation_dog_folder = "data/validation/dogs/" 

# create file list from the folders
cat_train_files = os.walk(train_cat_folder).__next__()[2]
cat_val_files = os.walk(validation_cat_folder).__next__()[2]
dog_train_files = os.walk(train_dog_folder).__next__()[2]
dog_val_files = os.walk(validation_dog_folder).__next__()[2]

# set the size of the training set (20000 images in total)
train_file_number_cat = 10000
train_file_number_dog = 10000

# if the directory not exist, then create a new folder
def check_dir(file_path):
    if not os.path.exists(file_path):
        os.makedirs(file_path)

# check and create folders for the data splitting
check_dir(train_cat_folder)
check_dir(train_dog_folder)
check_dir(validation_cat_folder)
check_dir(validation_dog_folder)

i = 0                           # counter for cat files
j = 0                           # counter for dog files

print("Splitting the data...")

# use os.walk to "walk through" all the pictures in train folder (downloaded from Kaggle)
for root, dirs, files in os.walk("train/", topdown=True):
    # get the name of each file
    for name in files:
        file_source = os.path.join(root, name)       # full path of the file

        if name[0] == "c":                          # check if picture name started with "c" -> cat
            if i < train_file_number_cat:
                shutil.move(file_source, os.path.join(train_cat_folder, name))
                i += 1
            else:
                shutil.move(file_source, os.path.join(validation_cat_folder, name))          

        if name[0] == "d":                          # check if the picture name started with "d" -> dog
            if j < train_file_number_dog: 
                shutil.move(file_source, os.path.join(train_dog_folder, name))
                j += 1
            else:
                shutil.move(file_source, os.path.join(validation_dog_folder, name))
print("Done!")

```

```python
print("File count:")
print("Cat traning - %s" % len(cat_train_files))
print("Cat validation - %s" % len(cat_val_files))
print("Dog traning - %s" % len(dog_train_files))
print("Dog validation - %s" % len(dog_val_files))
```

# Check point - Model Training

Here we are going to train 4 different models individually, based on the pre-trained models came with Keras:
    - VGG16
    - InceptionResNetV2
    - Xception
    - ResNet50
    
Then, we will create some essemble/stacking models by combining the 4 models above:
    - Model A -> VGG16, InceptionResNetV2, Xception, ResNet50
    - Model B -> InceptionResNetV2, Xception, ResNet50
    - Model C -> Xception, ResNet50
    - Model D -> InceptionResNetV2, Xception
    - Model E -> InceptionResNetV2, ResNet50

#### Import required libraries and shared variables

```python
import numpy as np
import pickle
import os
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Dropout, Flatten, Dense, Input, BatchNormalization, Activation
from keras.models import Model, Input
from keras import applications
from keras import optimizers
from keras.callbacks import EarlyStopping
%matplotlib inline

cat_train_files = os.walk("data/train/cats/").__next__()[2]
cat_val_files = os.walk("data/validation/cats/").__next__()[2]
dog_train_files = os.walk("data/train/dogs/").__next__()[2]
dog_val_files = os.walk("data/validation/dogs/" ).__next__()[2]

train_sample_size = len(cat_train_files) + len(dog_train_files)   
validation_sample_size = len(cat_val_files) + len(dog_val_files)

batch_size = 50
epochs = 50
```

## Training -> VGG based model

```python
learning_rate = 0.001
decay_rate = learning_rate/epochs

# create the model
model_input = Input(shape=(224, 224, 3))
base_model = applications.VGG16(include_top=False)
# base_model = applications.VGG16(include_top=False, weights="imagenet",input_shape=(224, 224, 3))
x = base_model(model_input)
x = Flatten()(x)
x = Dense(256, use_bias=False)(x)
x = BatchNormalization()(x)
x = Activation("relu")(x)
x = Dense(128, use_bias=False)(x)
x = BatchNormalization()(x)
x = Activation("relu")(x)
x = Dropout(0.5)(x)
x = Dense(1, use_bias=False)(x)
x = BatchNormalization()(x)
x = Activation("sigmoid")(x)

model = Model(inputs=model_input, outputs=x)

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.SGD(lr=learning_rate, decay=decay_rate, momentum=0.9),
              metrics=['accuracy']) 


# Add data augmentation
train_data_generator = ImageDataGenerator(
    rescale=1./255,
    rotation_range=180,
    shear_range=0.3,
    zoom_range=0.4,
    horizontal_flip=True,
    vertical_flip=True)

val_data_generator = ImageDataGenerator(rescale=1./255)


# read pictures from 'data/train' folder
train_data = train_data_generator.flow_from_directory(
    'data/train',            
    target_size=(224,224),
    batch_size=batch_size,   
    class_mode='binary')

# read pictures from 'data/validation' folder
validation_data = val_data_generator.flow_from_directory(
    'data/validation',          
    target_size=(224,224),
    batch_size=batch_size,   
    class_mode='binary')

print("Start training...")

history = model.fit_generator(
        train_data,
        steps_per_epoch=train_sample_size // batch_size,
        epochs=epochs,
        callbacks=[EarlyStopping(monitor='val_loss', patience=5)],      # add Early Stoppping for 5 epochs
        validation_data=validation_data,
        validation_steps=validation_sample_size // batch_size)

# save the trained model
model.save('model_vgg.h5')

# save the history
with open('history_vgg', 'wb') as file_pi:
        pickle.dump(history.history, file_pi)
        
print("Training finsihed!!")
```

## Check point - Create ensemble models

```python
# import required libraries
from keras.models import Model, Input, load_model
from keras.layers import Average
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers
```

```python
# generate the validation data
val_data_generator = ImageDataGenerator(rescale=1./255)
validation_data = val_data_generator.flow_from_directory(
    'data/validation',          
    target_size=(224,224),
    batch_size=50,   
    class_mode='binary')
```

```python
# load all the models saved previously

print("loading vgg16...")
model_vgg16 = load_model('model_vgg.h5')             # VGG16
print("loading irnv2...")
model_irnv2 = load_model('model_irnv2.h5')           # InceptionResNetV2
print("loading xception...")
model_xception = load_model('model_xception.h5')     # Xception
print("loading rn50...")
model_rn50 = load_model('model_rn50.h5')             # ResNet50

print("models loaded")
```

```python
# rename names of the models to avoid name conflict in ensemble model
model_vgg16.name = 'model_vgg16'
model_irnv2.name = 'model_irnv2'
model_xception.name = 'model_xception'
model_rn50.name = 'model_rn50'
```

```python
# define a reuseable function to create ensemble model
def ensemble(models, model_input):
    # collect outputs of models in a list
    yModels = [model(model_input) for model in models]
    
    # averaging outputs
    yAvg = Average()(yModels)
    
    # build model from same input and avg output
    model = Model(model_input, yAvg, name='ensemble_model')
    
    return model
```

```python
# create model combinations
model_combo_a = [model_vgg16, model_irnv2, model_xception, model_rn50]
model_combo_b = [model_irnv2, model_xception, model_rn50]
model_combo_c = [model_xception, model_rn50]
model_combo_d = [model_irnv2, model_xception]
model_combo_e = [model_irnv2, model_rn50]

# create model input shared by all models
model_input = Input(shape=(224, 224, 3))

# create essemble models
ensemble_model_a = ensemble(model_combo_a, model_input)
ensemble_model_b = ensemble(model_combo_b, model_input)
ensemble_model_c = ensemble(model_combo_c, model_input)
ensemble_model_d = ensemble(model_combo_d, model_input)
ensemble_model_e = ensemble(model_combo_e, model_input)

print('all models created')

# ensemble_model_a.summary()
```

```python
# compile all ensemble models
all_models = [ensemble_model_a, 
              ensemble_model_b, 
              ensemble_model_c, 
              ensemble_model_d, 
              ensemble_model_e]

for model in all_models:
    model.compile(loss='binary_crossentropy',
              optimizer=optimizers.SGD(lr=0.001),
              metrics=['accuracy']) 
    
print('All models compiled')
```

```python
X_test = []

# read and resize test images in order with OpenCV
for image in test_images:
    X_test.append(cv2.resize(cv2.imread(image), (224,224), interpolation=cv2.INTER_CUBIC))

# create data generater for the test set
# remember -> shuffle has be None, otherwise the ids won't match the prediction probability
test_datagen = ImageDataGenerator(rescale=1. / 255)
test_generator = test_datagen.flow(np.array(X_test), batch_size=50, shuffle=None)
```

```python
# define a function to generate .csv for submission

def getAnswerWith(prediction):
    
    # generate id numbers for the 'id' column in csv
    id = range(1, len(test_images) + 1)
    
    # create the date frame with two columns - id & label
    answer = pd.DataFrame({"id": id, "label":list(prediction)})
    
    # remove the '[' and ']' from datas in label column 
    answer['label'] = answer['label'].map(lambda x: str(x).lstrip('[').rstrip(']')).astype(float)
    
    return answer
```
